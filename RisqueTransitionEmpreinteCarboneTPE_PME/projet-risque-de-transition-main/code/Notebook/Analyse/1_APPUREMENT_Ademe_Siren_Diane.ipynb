**Un premier filtre : moins de 30% de missing & variable quasi constante**

```python
def recapNA(data):
      taux_na = data.isna().mean()*100
      var     = list(data.columns)
      dtype   = data.dtypes.values

      data = {
            'Variables' : var,
            'Dtype': dtype,
            'Taux_na': taux_na.values    
            }
      recap_na = pd.DataFrame(data)
      recap_na['na_>30%'] = recap_na.Taux_na > 30
      return recap_na
```

**Correction des aberrances par la méthode boxplot**

```python
def outlier(df):
   for col in df.columns:
      Q1 = df[col].describe()["25%"]
      Q3 = df[col].describe()["75%"]
      interQuart = Q3 - Q1
      MG = Q1-1.5*interQuart
      MD = Q3+1.5*interQuart

      cond1 = df[col] > MD
      cond2 = df[col] < MG
      ## Affectation
      df[col]=df[col].drop(df[col][cond1 | cond2].index, axis=0)
      
      df[cond1].fillna({col: MD}, inplace=True)
      df[cond2].fillna({col: MG}, inplace=True)

```

```python
## Taux d'aberrance
def taux_aberrance(df):
   gauche = []
   droite = []
   taux_gauche = []
   taux_droite = []
   try:
      for col in numeric_var:
         Q1 = df[col].describe()["25%"]
         Q3 = df[col].describe()["75%"]
         interQuart = Q3 - Q1
         gauche.append(Q1-1.5*interQuart)
         droite.append(Q3+1.5*interQuart)
         taux_gauche.append(
            (df[col] < Q1-1.5*interQuart).mean()*100
            )
         taux_droite.append(
            (df[col] > Q3+1.5*interQuart).mean()*100
         )
   except:
      print('Probleme avec ', col)
      
   aberrance ={
         'variables': numeric_var,
         'valeur_gauche': gauche,
         'valeur_droite' : droite,
         'taux_gauche' : taux_gauche,
         'taux_droite' : taux_droite
      }
   

   ab_df = pd.DataFrame(aberrance)
   return ab_df
```

**Imputation par KNN**

```python
## Liste des variables concernees
ASD_operating = ASD_copy.loc[:, ~ASD_copy.columns.isin(var_not_in_concern)].copy()

numeric_var = ASD_operating.select_dtypes('number').columns

var_objet = ASD_operating.select_dtypes('object').columns
```

```python
## Package pour l'imputaion
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OrdinalEncoder, MinMaxScaler
## Instanciation des outils de preprocessing
ord_encoder = OrdinalEncoder()
num_encoder = MinMaxScaler()
```

```python
## Encodage des variables catégorielles pour le KNN Imputer
def encode(data):
    '''function to encode non-null data and replace it in the original data'''
    #retains only non-null values
    nonulls = np.array(data.dropna())
    #reshapes the data for encoding
    impute_reshape = nonulls.reshape(-1,1)
    #encode date
    impute_ordinal = ord_encoder.fit_transform(impute_reshape)
    #Assign back encoded values to non-null values
    data.loc[data.notnull()] = np.squeeze(impute_ordinal)
    return data
```

```python
for columns in var_objet:
    encode(ASD_operating[columns])
```

```python
## Normalisation des variables numeric
ASD_operating[numeric_var] = pd.DataFrame(num_encoder.fit_transform(ASD_operating[numeric_var]), columns = ASD_operating[numeric_var].columns)
```

```python
## Application du KNN
imputer = KNNImputer(n_neighbors=5)
ASD_operating = pd.DataFrame(imputer.fit_transform(ASD_operating),columns = ASD_operating.columns)
```
