Nous comparons les résultats des modèles classiques aux modèles top-down. <p>
Il s'agit entre d'autres termes de comparer les erreurs de prédiction des différents modèles. Pour cela, nous avons d'abord fait une comparaison des erreurs quadratiques moyennes des différents modèles. Ensuite, pour apprécier la différence significative en termes d'EQM, nous réalisons ici un test de différence de moyennes (i.e test de différences d'EQM) sur les échantillons test des différents modèles.

```python
## Importation des packages nécessaires
import os
import pandas as pd
import numpy as np
from scipy import stats ## Pour le test 
```

```python
# Définition du repertoire de travail
os.chdir("D:/ENSAI/2A/Projet Statistique/projet-risque-de-transition/")  # Mettez ic le repertoire de vos bases
```

```python
# Importation des bases résultat des modélisations (avec y_true et y_pred)
df_compare = pd.read_csv("bdd/Base_finale/df_comparaison.csv")
df_classique = pd.read_csv("bdd/Base_finale/df_predict_reg.csv")
df_ratio = pd.read_csv("bdd/Base_finale/temp_comparaison.csv")
```

### On fusionne les bases du modèle ratio et modèle filière, comme elles portent sur les mêmes individus de test.

```python
df_ratio['fusion'] =  df_ratio.index
df_compare['fusion'] = df_compare.index
```

```python
df_compare = df_compare.merge(df_ratio, on = 'fusion', how='left')
```

```python
df_classique
```

#### Calcul d'erreurs de prédiction (eqm)

```python
df_classique = df_classique.assign(mse_classique = (df_classique["y_true_reg"] - df_classique["y_classique"])**2)
df_compare = df_compare.assign(mse_filiere = (df_compare["y_true"] - df_compare["y_filiere"])**2)
#df_compare = df_compare.assign(mse_ratio = (df_compare["y_ratio_test"] - df_compare["y_ratio_predict"])**2)
```

```python
df_compare.isna().sum()
```

## Test de différence de moyennes

Nous supposons d'abord l'égalité des variances et après l'inégalité.

```python
## Test de différence de moyenne
stat, p_value = stats.ttest_ind(df_classique.mse_classique, df_compare.mse_filiere, equal_var=True)
print ("====================================================\n",
"Modèle classique et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_ind(df_classique.mse_classique, df_compare.mse_ratio, equal_var=True)
print ("====================================================\n",
"Modèle classique et modèle ratio \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_ind(df_compare.mse_ratio, df_compare.mse_filiere, equal_var=True)
print ("====================================================\n",
"Modèle ratio et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)
```

TEST POUR DES VARIANCES INEGALES

```python
## Test de différence de moyenne
stat, p_value = stats.ttest_ind(df_classique.mse_classique, df_compare.mse_filiere, equal_var=False)
print ("====================================================\n",
"Modèle classique et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_ind(df_classique.mse_classique, df_compare.mse_ratio, equal_var=False)
print ("====================================================\n",
"Modèle classique et modèle ratio \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_ind(df_compare.mse_ratio, df_compare.mse_filiere, equal_var=False)
print ("====================================================\n",
"Modèle ratio et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)
```

## Test de différence de moyennes pour des échantillons appariés

```python
## Test de différence de moyenne
stat, p_value = stats.ttest_rel(df_classique.mse_classique, df_compare.mse_filiere)
print ("====================================================\n",
"Modèle classique et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_rel(df_classique.mse_classique, df_compare.mse_ratio)
print ("====================================================\n",
"Modèle classique et modèle ratio \n  Statistique: ", stat, "\n  p-valeur : ", p_value)

stat, p_value = stats.ttest_rel(df_compare.mse_ratio, df_compare.mse_filiere)
print ("====================================================\n",
"Modèle ratio et modèle filière \n  Statistique: ", stat, "\n  p-valeur : ", p_value)
```

```python
## Fonction pour test des tests unilatéraux ou bilatéraux
def welch_ttest(x1, x2,alternative):
    n1 = x1.size
    n2 = x2.size
    m1 = np.mean(x1)
    m2 = np.mean(x2)
    v1 = np.var(x1, ddof=1)
    v2 = np.var(x2, ddof=1)
    t = (m1 - m2) / np.sqrt(v1 / n1 + v2 / n2)
    df = (v1 / n1 + v2 / n2)**2 / (v1**2 / (n1**2 * (n1 - 1)) + v2**2 / (n2**2 * (n2 - 1)))
    if alternative == "equal":
        p = 2 * t.cdf(-abs(t), df)
    if alternative == "lesser":
        p = 1-t.cdf(abs(t), df)
    if alternative == "greater":
        p = t.cdf(abs(t), df)
    return t, df, p
```

## Intervalle de confiance

On calcule ici l'IC de l'EQM de chacun des modèles.

```python

print("===================================\n mse_classique\n", np.mean(df_classique["mse_classique"]), " et ",
    "IC à 95: [",np.mean(df_classique["mse_classique"]) - 1.96*np.std(df_classique["mse_classique"])/np.sqrt(df_classique.shape[0]),
    ", ",np.mean(df_classique["mse_classique"]) + 1.96*np.std(df_classique["mse_classique"])/np.sqrt(df_classique.shape[0]),"] \n")
```
