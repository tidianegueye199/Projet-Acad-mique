## Base Industrie manufacturière et Construction

Pour ce secteur, on distingue deux grands sous-secteurs: Industrie manufacturière d'une part et Construction d'autre part. Donc nous créons deux bases pour les deux pour des besoins avenirs.

### Construction

```python
## Récupération des données du sous-secteur construction
def  donnees(ligne):
    # Chargement de la base
    workbook = xlrd.open_workbook("Citepa_Industrie-manufacturiere-construction_Secten_ed2021-d.xlsx")
    # Precision de la feuille de données
    worksheet = workbook.sheet_by_name("Industrie-Emissions")
    # Initialiser la liste où mettre les données
    donnees = []
    
    # On parcourt les colonnes pour récupérer les données
    for curr_col in range(3 + 63 -30,61 + 6, 1):
        # ligne sera entrée en parmètre pour permettre la précision des types de GES
        data = worksheet.cell_value(ligne, curr_col)
        donnees.append(data)

    return donnees
```

```python
## On met en entrée la ligne correspondant aux données de chaque type de GES pour la construction
# Et récuperer toutes les valeurs sur toutes les années
annee = donnees(9)
COdeux = donnees(25)
COdeuxe = donnees(11)
CHquatre = donnees(39)
NdeuxO = donnees(53)
HFCs = donnees(67)
PFCs = donnees(81)
SFsix = donnees(95)
NFtrois = donnees(109)
Gaz_fluores = donnees(123)

## On crée un dataframe des émissions du secteur de construction avec les années en ligne
secten_construct = pd.DataFrame([annee, COdeux, COdeuxe, CHquatre,NdeuxO,HFCs,PFCs,SFsix,NFtrois,Gaz_fluores],
index = ["annee", "CO2", "CO2e", "CH4","N2O","HFC","PFC","SF6","NF3","Gaz_fluores"])

# Pour avoir la même structure que les bases précédentes, on transpose cette base pour avoir comme variables 
# les différents GES 
secten_construction = secten_construct.transpose()
# Création de la variable secteur précisant le secteur de construction
secten_construction["secteur"] = "construction"
```

```python
## Aperçu sur la base construction
secten_construction.head()
```

### Industrie manufacturière

Pour créer la base du sous-secteur industrie manufacturière, nous allons d'abord créer la base de tout le secteur (Industrie + construction), ensuite enlever les émissions du sous-secteur construction.

```python
# Chargement la base des données
# Dans la méthode, data_scrap.chemin([Nom fichier excel], [La feuille de données])
worksheet = data_scrap.chemin("Citepa_Industrie-manufacturiere-construction_Secten_ed2021-d.xlsx",
"Industrie-Emissions")

# Recuperation des données GES par type et année
secten_industrie = data_scrap.data_frame(worksheet, 496, 4)
# Création de la variable spécifiant le secteur
secten_industrie["secteur"] = "indus_manu"
secten_industrie.head()
```

On renomme les variables pour les deux bases

```python
# renommer pour la base complète
dic_i = ["annee","CO2i", "CO2ei","CH4i", "N2Oi",
"HFCi","PFCi","SF6i","NF3i","Gaz_fluoresi", "secteur"]
# renommer pour la base construction
dic_c = ["annee","CO2c", "CO2ec","CH4c", "N2Oc",
"HFCc","PFCc","SF6c","NF3c","Gaz_fluoresc", "secteur"]

# Affection des nouveaux noms pour chacune des bases
secten_industrie.set_axis(dic_i, axis='columns', inplace= True)
secten_construction.set_axis(dic_c, axis='columns', inplace= True)
```

Etant des noms de variables différents, nous allons fusionner les deux bases avec comme variable de fusion *_annee_*.

```python
# Fusion de deux bases
df =secten_industrie.merge(secten_construction, on='annee',validate='1:1')
```

```python
## Faire la différence pour chaque GES pour obtenir les valeurs pour le sous-secteur industrie manufacturière
df = df.assign(CO2 = df.CO2i - df.CO2c,
CO2e = df.CO2ei - df.CO2ec,
CH4 = df.CH4i - df.CH4c,
N2O = df.N2Oi - df.N2Oc,
HFC = df.HFCi - df.HFCc,
PFC = df.PFCi - df.PFCc,
SF6 = df.SF6i - df.SF6c,
NF3 = df.NF3i - df.NF3c,
Gaz_fluores = df.Gaz_fluoresi- df.Gaz_fluoresc,
secteur = df.secteur_x)
```

Pour avoir uniquement les données du sous-secteur industrie manufacturière, on garde seulement les variables créées précédemment.

```python
## Création de la base sous-secteur industrie manufacturière
secten_industrie = df[["annee", "CO2", "CO2e", "CH4","N2O","HFC","PFC","SF6","NF3","Gaz_fluores", "secteur"]]

# pour la base du sous-secteur construction, on ramène les noms des variables initiaux
dic_c = ["annee","CO2", "CO2e","CH4", "N2O",
"HFC","PFC","SF6","NF3","Gaz_fluores", "secteur"]
secten_construction.set_axis(dic_c, axis='columns', inplace= True)
secten_industrie.info()
```

## Base Residentiel-Tertiaire

Nous réalisons le même traitement pour la base du secteur Résidentiel-Tertiaire en distinguant les deux sous-secteurs: Tertitaire et résidentiel.

```python
# Chargement la base des données globale
# Dans la méthode, data_scrap.chemin([Nom fichier excel], [La feuille de données])
worksheet = data_scrap.chemin("Citepa_Batiments-Residentiel-Tertiaire_Secten_ed2021-d.xlsx",
"RT-Emissions")

secten_rt = data_scrap.data_frame(worksheet, 661, 4)
secten_rt["secteur"] = "RT"
secten_rt.head()
```

### Tertiaire

```python
## Récupération des données 
## Création d'une fonction permettant de parcourir les colonnes d'observation pour chaque ligne entrée en paramètre.
## La ligne correspond à un type de GES et ses valeurs d'émissions au cours des années
def  donnees(ligne):
    # Chargement de la base brute
    workbook = xlrd.open_workbook("Citepa_Batiments-Residentiel-Tertiaire_Secten_ed2021-d.xlsx")
    # Précison de la feuille de données
    worksheet = workbook.sheet_by_name("RT-Emissions")
    donnees = []
    ## Boucle pour récuperer les données sur plusieurs années
    for curr_col in range(3 + 63 -30,61 + 6, 1):
        data = worksheet.cell_value(ligne, curr_col)
        donnees.append(data)

    return donnees
```

Liste des données tous les types de GES pour le sous-secteur Tertiaire et création de la base Tertiaire

```python
# Liste des données tous les types de GES pour le sous-secteur (SUR LES ANNEES)
annee = donnees(9)
COdeux = donnees(42)
COdeuxe = donnees(23)
CHquatre = donnees(61)
NdeuxO = donnees(80)
HFCs = donnees(99)
PFCs = donnees(118)
SFsix = donnees(137)
NFtrois = donnees(156)
Gaz_fluores = donnees(175)

secten_tert = pd.DataFrame([annee, COdeux, COdeuxe, CHquatre,NdeuxO,HFCs,PFCs,SFsix,NFtrois,Gaz_fluores],
index = ["annee", "CO2", "CO2e", "CH4","N2O","HFC","PFC","SF6","NF3","Gaz_fluores"])
secten_tertiaire = secten_tert.transpose()
secten_tertiaire["secteur"] = "tertiaire"
```

### Résidentiel

Liste des données tous les types de GES pour le sous-secteur résidentiel et création de la base Résidentiel

```python
# Bases pour résidentiel
annee = donnees(9)
COdeux = donnees(36)
COdeuxe = donnees(17)
CHquatre = donnees(55)
NdeuxO = donnees(74)
HFCs = donnees(93)
PFCs = donnees(112)
SFsix = donnees(131)
NFtrois = donnees(150)
Gaz_fluores = donnees(169)

secten_resid = pd.DataFrame([annee, COdeux, COdeuxe, CHquatre,NdeuxO,HFCs,PFCs,SFsix,NFtrois,Gaz_fluores],
index = ["annee", "CO2", "CO2e", "CH4","N2O","HFC","PFC","SF6","NF3","Gaz_fluores"])
secten_residentiel = secten_resid.transpose()
secten_residentiel["secteur"] = "residentiel"
```

# Fusion des différentes bases

Nous nous sommes assuré d'avoir la même structure pour toutes les bases et une variable précisant le secteur. Nous pouvons aisément concatener toutes ces bases pour obtenir notre base finale sur les émissions de GES des différents secteurs selon Citepa.

```python
# Création de la base finale à l'aide de la méthode concat de pandas
base_secten = pd.concat([secten_energie, secten_industrie, secten_utcatf, secten_emi_naturelles,
    secten_agriculture, secten_transport, secten_dechets, secten_residentiel,
    secten_tertiaire, secten_construction], ignore_index=True)
```

```python
# Qualité de la fusion
base_secten.info()
```

## Correction variable année

Dans les données, les émissions de 2020 sont estimées pour tous les secteurs. Nous corrigons la modalité indiquant cette année à l'aide du code ci-dessous.

```python
# creation de la variable annee
for i in range(base_secten.shape[0]):
    if base_secten["annee"][i] == '2020 (e) ':
        base_secten["annee"][i] = 2020

base_secten = base_secten.astype({"annee": int}) 
```

```python
## Head de la base finale
base_secten.head()
```

```python
# Vérification des valeurs manquantes
base_secten.isna().sum()
```

Nous allons calculer le total en le rapportant en Mégatonne, donc diviser les valeurs des GES autres que CO2 par 1000. <p>
<br> Total_ges = somme pondérée des différents gaz (CO2, CH4, etc.) </br>

```python
# Calcul du total d'émissions de GES
base_secten["total_ges"] = base_secten["CO2"] + base_secten["CO2e"] +\
    (base_secten["CH4"] + base_secten["N2O"] + base_secten["HFC"] + \
        base_secten["PFC"] + base_secten["SF6"] + base_secten["NF3"] + \
            base_secten["Gaz_fluores"])/1000

```
